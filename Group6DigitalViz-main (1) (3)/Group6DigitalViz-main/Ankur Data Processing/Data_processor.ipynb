```python
import pandas as pd
```

Datasets pulled and customised from the World Bank's databank: https://databank.worldbank.org/reports.aspx?source=2&series=FP.CPI.TOTL.ZG&country=#

Data.csv: Pulled from databank with multiple economic indicators, renamed, and then filtered to contain entries for "Getting electricity: Cost", for select countries.

Inflation_data.csv: Pulled from databank to contain only "Inflation, consumer prices (annual %)" field, renamed and then filtered for selct countries.

```python
data = pd.read_csv('data.csv')
data
```

```python
data['Indicator'].unique()
```

```python
data = data[data['Indicator'] == 'Getting electricity: Cost']
data.index=data['Country Name']
data = data[['2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020']]
data
```

```python
data.loc['World'] = data.mean()
data
```

```python
select_countries = data[data.index.isin(['World', 'China', 'India', 'United Kingdom', 'United States', 'Germany'])]
select_countries
```

```python
inflation_data = pd.read_csv('Inflation_Data.csv')
inflation_data.columns
```

```python
inflation_data.index=inflation_data['Country Name']
inflation_data = inflation_data[['2008 [YR2008]', '2009 [YR2009]', '2010 [YR2010]', '2011 [YR2011]', '2012 [YR2012]',
       '2013 [YR2013]', '2014 [YR2014]', '2015 [YR2015]', '2016 [YR2016]',
       '2017 [YR2017]', '2018 [YR2018]', '2019 [YR2019]', '2020 [YR2020]', '2021 [YR2021]']]
inflation_data.columns = ['2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021']
inflation_data.dropna(inplace=True)
inflation_data
```
